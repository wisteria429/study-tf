```python
def resize(data, w, h):
    img = Image.new("L", (w, h))
    pix = img.load()
    for i in range(h):
        for j in range(w):
            pix[i, j] = int(data[j][i])
    img2 = img.convert("RGB")
    img2 = img2.resize((224, 224))
    return img2
```

```python
def save(img, save_path, filename):
    path = save_path + "/" + filename + ".jpg"
    print(path)
    img.save(path)
```

```python
def resizeAndSave(x, y, prefix):
    size, w_size, h_size = x.shape
    for i in range(10000): 
        dirName = prefix + "/"+ str(y[i])
        if os.path.isdir(dirName) is False:
            os.makedirs(dirName)
        
        img = resize(x[i], w_size, h_size)
        
        save(img, dirName,str(i))
        
```

```python
resizeAndSave(x_train, y_train, "TRAIN")
#resizeAndSave(x_test, y_test, "TEST")
```
